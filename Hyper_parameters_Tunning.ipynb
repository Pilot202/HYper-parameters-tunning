**Hyperparameters tunning involves the adjusting the configuration of a model known as hyperparameters. This help to improve the accuracy of the model or to decrease the errors by finding the best hyperparameters combination.

```python
#importing Necessary libraries and Data
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_digits
```

```python
digits= load_digits()
dir(digits)
```

```python
digits.target
```

```python
df=pd.DataFrame(digits.data, digits.target)
df['target']= digits.target
df
```

```python
X= df.drop('target', axis= 1)
X
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test= train_test_split(X,df.target, test_size=0.2)
```

**Using Gridsearch method to find the best model and hyperparameters for our model

```python
from sklearn.model_selection import GridSearchCV

clf= GridSearchCV(svm.SVC(gamma='auto'), {
    'C': [1,20,10],
    'kernel': ['linear', 'rbf']
}, cv=5, return_train_score=False)
clf.fit(X_train, y_train)
clf.cv_results_
```

**The printed Information from The above cell can be put into a Dataframe for easy comprehension

```python
table= pd.DataFrame(clf.cv_results_)
table
```

```python
df=table[['param_C','param_kernel','params','mean_test_score']]
df
```

```python
clf.best_params_
```

```python
clf.best_score_
```

**In other to reduce the number of Iteration so as to save Computational Cost, Developers make use of RANDOMSEARCH  Method, which 
take few random parameters and use it to select the best Hyperparameters.

```python
from sklearn.model_selection import RandomizedSearchCV
rms= RandomizedSearchCV(svm.SVC(gamma= 'auto'), {
    'kernel': ('linear', 'rbf'),
    'C':(1,10,20,30)}, cv=5, return_train_score= False, n_iter=3)
rms.fit(X_train,y_train)
rms.cv_results_
```

```python
df2= pd.DataFrame(rms.cv_results_)
df2
```

```python
df2[['param_kernel','param_C','params','mean_test_score']]
```

For testing with several model with diferent parameters, we can make use of the following simple method

```python
model_chars= {
    'SVC': {
        'model' :svm.SVC(),
        'param': {
            'kernel': ['rbf','linear'],
        'C': [1,10,20]}
           },
    'LogisticRegression': {
        'model' :LogisticRegression(solver='liblinear',multi_class='auto'),
        'param': {
        'C': [1,10,20]},
           },
    'RandomForest': {
        'model' :RandomForestClassifier(),
        'param': {
          
        'n_estimators': [40,30,20]}
    },
# #     'MultinomialNB': {
        
#         'model' :MultinomialNB(),
#     },
    "DecisionTree":{
        'model' :DecisionTreeClassifier(),
        'param': {
            'max_depth': [5,8,10]
        }
    }
}
```

```python
scores= []
for model_name, mp in model_chars.items():
    ggs= GridSearchCV(mp['model'], mp['param'], cv=5, return_train_score= False)
    ggs.fit(X_train, y_train)
    scores.append({
        'model': model_name,
        'parameters': ggs.best_params_,
        'best_scores': ggs.best_score_
    })
scores
```

```python
df3=pd.DataFrame(scores,columns=['model', 'parameters', 'best_scores'])
df3
```

```python
score= []
for model_name, mp in model_chars.items():
    ggs= RandomizedSearchCV(mp['model'], mp['param'], cv=5,n_iter=3, return_train_score= False)
    ggs.fit(X_train, y_train)
    score.append({
        'model': model_name,
        'parameters': ggs.best_params_,
        'best_scores': ggs.best_score_
    })
score
```

```python
df4= pd.DataFrame(score, columns=['model', 'parameters', 'best_scores'])
```

from the table above, SVC with kernel= rbf and C=20 has the best_score. so therefore making it the most suitablemodel for our predictions.

```python
model= svm.SVC(kernel='rbf', C=20)
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```
